```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
X = [[130], [250], [190], [300], [210], [220], [170]]
y = [16.3, 10.2, 11.1, 7.1, 12.1, 13.2, 14.2]
```

```python
regr=linear_model.LinearRegression()

regr.fit(X, y)
```

```python
coef = regr.coef_
intercept = regr.intercept_
print('y =',coef, '* X + ', intercept )

score = regr.score(X, y) 
print(score)
```

```python
plt.scatter(X, y, color='black')

y_pred = regr.predict(X)
print('예측된 결과 = ', y_pred)

plt.plot(X, y_pred, linewidth=3)

new_data = [[270]]
new_pred = regr.predict(new_data)

plt.scatter(new_data, new_pred, color='red')
print('예측된 결과 = ', new_pred)
plt.show()
```
