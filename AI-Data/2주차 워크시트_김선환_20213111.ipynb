# 3. CCTV 현황 데이터 정리하기

```python
#CCTV_Seoul 데이터의 0번째 열이름을 '구분' -> '구별'로 변경 (columns={'구분' : '구별'})
CCTV_Seoul.rename(columns = {'구분' : '구별'}, inplace = True)
CCTV_Seoul.head()
```

```python
#다른 데이터들과 년도를 동일하게 맞추기 위해 '2011년 이전과 2020년' 열 삭제 (columns=['2011년 이전'])
CCTV_Seoul.drop(columns = ['2011년 이전'], axis = 1, inplace = True)
CCTV_Seoul.drop(columns = ['2020년'], axis = 1, inplace = True)
CCTV_Seoul.head()
```

```python
#'2017년 이전' 열 추가  2017이전 = 총계-(2017년+2018년+2019년)
CCTV_Seoul['2017년 이전'] = CCTV_Seoul['총계'] - (CCTV_Seoul['2017년']+CCTV_Seoul['2018년']+CCTV_Seoul['2019년'])
CCTV_Seoul.head()
```

```python
#불필요한 년도의 열('2011년 이전', '2011년', '2012년', '2013년', '2014년','2015년','2016년') 삭제
CCTV_Seoul.drop(columns=['2011년', '2012년', '2013년', '2014년', '2015년','2016년'], axis=1, inplace=True)
CCTV_Seoul.head()
```

```python
#불필요한 행(index=0) 삭제
CCTV_Seoul.drop(index = 0, axis = 0, inplace = True)
CCTV_Seoul.head()
```

```python
 #기존의 인덱스는 버리고 새로 설정
CCTV_Seoul.reset_index(drop = True, inplace = True)
CCTV_Seoul.head()
```

# 4. CCTV 현황데이터 자세히 보기

### 4-1. CCTV의 전체 개수가 가장 적은/가장 많은 상위 5개 구는 어디일까?

```python
#데이터 '총계'열을 기준으로 오름차순 정렬하기 (by='총계', ascending = True)

CCTV_Seoul.sort_values(by = '총계', ascending = True).head()
```

```python
#데이터 '총계'열을 기준으로 내림차순 정렬하기 (by='총계', ascending = False)

CCTV_Seoul.sort_values(by = '총계', ascending = False).head()
```

### 4-2. 최근 3년간 CCTV 증가율을 계산하여 '최근증가율' 을 알아보자.

```python
#최근 3년간 CCTV 수를 더하고 2017년 이전 CCTV수로 나눠서 최근 3년간 CCTV 증가율 계산

CCTV_Seoul['최근증가율']= (CCTV_Seoul['2017년'] + CCTV_Seoul['2018년'] + CCTV_Seoul['2019년'])/CCTV_Seoul['2017년 이전']*100
CCTV_Seoul.head()
```

### 4-3. '최근증가율' 이 가장 높은 상위 5개구는 어디일까?

```python
#'최근증가율'로 내림차순 정렬
CCTV_Seoul.sort_values(by = '최근증가율', ascending = False).head()
```
