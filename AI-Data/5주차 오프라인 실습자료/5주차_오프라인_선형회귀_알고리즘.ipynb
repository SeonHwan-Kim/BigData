```python
################################################################################
# 1. 선형회귀를 위한 라이브러리 준비하기
# (1) 라이브러리 불러오기
# (2) 선형회귀 준비하기
# (3) 학습데이터 준비하기
# (4) 선형회귀 생성하기
################################################################################
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model

#학습데이터과 정답데이터 준비
X = [[164],[179],[162],[170]]
y = [53, 63, 55, 59]

#선형회귀 준비하기
regr=linear_model.LinearRegression()

#선형회귀 생성하기
regr.fit(X, y)
```

```python
################################################################################
# 2. 만들어진 선형회귀 분석 결과 확인하기_1
# (1) 계수확인하기
# (2) 절편 확인하기
# (3) 차트로 생성해보기
################################################################################

# y = ax + b 에서 선형회귀 결과에서 도출된 a(계수)와 b(절편) 확인하기
coef = regr.coef_
intercept = regr.intercept_
print('y =',coef, '* X + ', intercept )

# X에 대해 y를 예측하는데 얼마나 적합한지 알아보기
# 1에 가까울수록 선형관계에 가깝다고 할 수 있음
score = regr.score(X, y)   

#학습데이터인 X와 결과데이터 y를 이용하여 scatter 차트 그려보기
plt.scatter(X, y, color='black')
```

```python
################################################################################
# 3. 선형회귀분석 결과 확인하기_2
# (1) 원본데이터(학습데이터터)를 이용하여 예측한 후 정답 데이터와 비교해보기
# (2) 선형회귀 분석 결과를 차트로 그려보기
###############################################################################

#학습데이터를 사용하여 결과 데이터에 얼마나 근접한지 확인해 보기
y_pred = regr.predict(X)
print('예측된 결과 = ', y_pred)

#원본 데이터를 예측된 데이터로 사용하여 line 차트 그려보기
plt.plot(X, y_pred, linewidth=3)
```

```python
################################################################################
# 4. 분석된 선형회귀를 이용하여 예측해보기
# (1) 새로운 학습데이터 준비하기
# (2) 새로운 학습데이터로 예측하기
# (3) 선형회귀선과의 거리를 차트로 확인해보기
################################################################################

#새로운 학습데이터를 사용하여 결과 데이터 확인하고 선형회귀선과 얼마나 가까운지 확인해보기
new_data = [[180], [185], [166], [179]]
new_pred = regr.predict(new_data)

plt.scatter(new_data, new_pred, color='red')
plt.show()
```
