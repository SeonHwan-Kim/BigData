```python
#1. 의사결정트리를 생성하기 위한 라이브러리 준비하기
from sklearn import tree
```

```python
# 2. 학습데이터와 정답 데이터 준비하기
x = [[2,0,1], [2,0,0], [4,1,1], [4,0,1]]
y = [1, 2, 0, 0] 

# 3. 의사결정트리 준비하기
clf = tree.DecisionTreeClassifier()

# 4. 학습데이터와 정답 데이터를 이용하여 의사결정트리 생성하기
clf = clf.fit(x, y)
```

```python
# 5. 생성된 의사결정트리를 그래프로 생성하기
from sklearn.tree import export_graphviz
export_graphviz(clf, 
                out_file='tree.dot', 
                class_names=['자동차', '오토바이', '자전거'], 
                impurity=False, 
                filled=True)
```

```python
# 6. 생성된 의사결정트리를 그래프로 출력하기

import graphviz
with open('tree.dot') as f : 
  dot_graph = f.read()
display(graphviz.Source(dot_graph))
```

```python
# 6. 새로운 학습데이터로 예측 결과 확인하기
# 0: 자동차
# 1: 오토바이
# 2:자전거
clf.predict([[2,1,1]])   

clf.predict([[3,1,0]])
```
